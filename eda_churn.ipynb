# Junção dos DataFrames

```python
len(df_clients)
```

```python
df_services.rename(columns={'customerID': 'IDCliente'}, inplace=True)
```

```python
df_services.info()
```

```python
df_temp = df_clients.merge(df_services, on=['IDCliente'])
```

```python
df_temp.info()
```

```python
df_temp.head()
```

```python
df_churn_temp = df_temp.merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
df_churn_temp.info()
```

```python
df_churn_temp.head()
```

```python
df_churn = df_clients.merge(df_services, on=['IDCliente']).merge(df_contracts, left_on='IDCliente', right_on='customerID')
```

```python
df_churn.info()
```
